```python
import json
import urllib.request
import urllib.parse
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
```

```python
# CSV 파일로드 
urdata = pd.read_csv('user_crowing_dropna.csv', encoding = 'utf-8')
chid = pd.DataFrame(columns=['urname','characterId'])

# API에서 유저 고유 번호 추출
for i in range(0 , len(urdata)) :
    # svname = 서버이름, urname = 닉네임
    svname = urdata.iloc[i, 0]
    urname = urdata.iloc[i, 1]

    # 닉네임 URL 변환
    urname_url = urllib.parse.quote(urname)
    
    # 아이디 검색 API
    requestData = requests.get(f'https://api.neople.co.kr/df/servers/{svname}/characters?characterName={urname_url}&apikey=M23RMaIB2JlKPyRWYRBKDb6UtE3NDTrS')
    if requestData.status_code == 200 :
        jsonData = requestData.json()
        api_result = pd.DataFrame(jsonData['rows'])
        # 닉네임 이상한거 Null값 처리
        if api_result.empty :
            characterId_filter = pd.DataFrame({'urname' : [urname], 'characterId' : None})
        else :
            characterId_filter = pd.DataFrame({'urname' : api_result.characterName, 'characterId' : api_result.characterId})
    else :
        characterId_filter = pd.DataFrame({'urname' : [urname], 'characterId' : None})
    chid = pd.concat([chid, characterId_filter])

chid.to_csv('2.1_usercode_result_backup.csv', encoding = 'euc-kr')
```
