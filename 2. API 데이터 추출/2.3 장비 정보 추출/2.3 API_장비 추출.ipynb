```python
import json
import urllib.request
import urllib.parse
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
from time import sleep
```

```python
# 데이터 로드
urdata = pd.read_csv('2.2_Status_result.csv', encoding = 'euc-kr')

# 빈 데이터 프레임 생성
user_info = pd.DataFrame(columns = ['urname', 'weapon', 'jacket', 'shoulder', 'pants', 'shoes', 'waist', 'amulet', 'wrist', 'ring', 'support', 'magicstone', 'earring','fusion ring'])

for i in range(202104, len(urdata)) :
    urname = urdata.iloc[i,0]
    svname = urdata.iloc[i,1]
    chid = urdata.iloc[i,3]
    requestData = requests.get(f'https://api.neople.co.kr/df/servers/{svname}/characters/{chid}/equip/equipment?apikey=M23RMaIB2JlKPyRWYRBKDb6UtE3NDTrS')
    jsonData = requestData.json()
    if requestData.status_code == 200 :
        api_result = pd.json_normalize(jsonData)
        user_equipment = pd.json_normalize(jsonData , record_path = 'equipment')
        try:
            weapon = user_equipment.loc[0, 'itemName']
            jacket = user_equipment.loc[2, 'itemName']
            shoulder = user_equipment.loc[3, 'itemName']
            pants = user_equipment.loc[4, 'itemName']
            shoes = user_equipment.loc[5, 'itemName']
            waist = user_equipment.loc[6, 'itemName']
            amulet = user_equipment.loc[7, 'itemName']
            wrist = user_equipment.loc[8, 'itemName']
            ring = user_equipment.loc[9, 'itemName']
            fusion_ring = user_equipment.loc[9, 'upgradeInfo.itemName']
            support = user_equipment.loc[10, 'itemName']
            magicstone = user_equipment.loc[11, 'itemName']
            earring = user_equipment.loc[12, 'itemName']

            user_equipment_name = pd.DataFrame({
                        'urname' : [urname],
                        'weapon' : [weapon],
                        'jacket' : [jacket],
                        'shoulder' : [shoulder],
                        'pants' : [pants],
                        'shoes' : [shoes],
                        'waist' : [waist],
                        'amulet' : [amulet],
                        'wrist' : [wrist],
                        'ring' : [ring],
                        'support' : [support],
                        'magicstone' : [magicstone],
                        'earring' : [earring],
                        'fusion ring' : [fusion_ring]
                        })
            user_info = pd.concat([user_info, user_equipment_name])
        except KeyError :
            user_equipment_name = pd.DataFrame({
                        'urname' : [urname],
                        'weapon' : None,
                        'jacket' : None,
                        'shoulder' : None,
                        'pants' : None,
                        'shoes' : None,
                        'waist' : None,
                        'amulet' : None,
                        'wrist' : None,
                        'ring' : None,
                        'support' : None,
                        'magicstone' : None,
                        'earring' : None,
                        'fusion ring' : None
                        })
            user_info = pd.concat([user_info, user_equipment_name])
    else :
        user_equipment_name = pd.DataFrame({
                        'urname' : [urname],
                        'weapon' : None,
                        'jacket' : None,
                        'shoulder' : None,
                        'pants' : None,
                        'shoes' : None,
                        'waist' : None,
                        'amulet' : None,
                        'wrist' : None,
                        'ring' : None,
                        'support' : None,
                        'magicstone' : None,
                        'earring' : None,
                        'fusion ring' : None
                        })
        user_info = pd.concat([user_info, user_equipment_name])
    print(f'{i}번째까지')
    sleep(0.1)

total_result_equipment = pd.merge(urdata, user_info, on = 'urname')
total_result_equipment.to_csv('2.3 result.csv', encoding='euc-kr')
```
